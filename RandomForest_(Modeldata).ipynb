```python
md1_x = md1.iloc[:, 0:13]     #independent variables
md1_y = md1.iloc[:, -1]       #dependent varaiable
md1_x.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
md1_x_train,md1_x_test,md1_y_train,md1_y_test = train_test_split(md1_x,md1_y,test_size = .2, random_state = 101)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state = 2)
```

```python
md1_x_over,md1_y_over = sm.fit_sample(md1_x_train, md1_y_train.ravel())
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(md1_x_over, md1_y_over)
```

```python
pred = rfc.predict(md1_x_test)
pred
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(pred, md1_y_test)
matrix
```

```python
acc = matrix.diagonal().sum()/matrix.sum()
acc
```
