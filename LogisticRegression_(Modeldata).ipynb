# Data Cleaning

```python
md1.isnull().sum()
```

```python
md1.purpose = md1.purpose.fillna('ME_PROFESSIONAL')
md1.job = md1.job.fillna('STUDENT')
md1.product_id = md1.product_id.fillna('APRO')
md1.download_attempts = md1.download_attempts.fillna(4)
md1.product_platforms = md1.product_platforms.fillna('WIN')
md1.business_segment = md1.business_segment.fillna('INDIVIDUAL')
md1.machines = md1.machines.fillna(1)
md1.isnull().sum()
```

```python
md1.skill = le.fit_transform(md1.skill)                   
md1.job = le.fit_transform(md1.job)
md1.purpose = le.fit_transform(md1.purpose)
md1.target_group = le.fit_transform(md1.target_group)
md1.business_group = le.fit_transform(md1.business_group)
md1.business_segment = le.fit_transform(md1.business_segment)
md1.signup_source = le.fit_transform(md1.signup_source)
md1.country_code = le.fit_transform(md1.country_code)
md1.product_id = le.fit_transform(md1.product_id)
md1.product_platforms = le.fit_transform(md1.product_platforms)
md1.signup_category = le.fit_transform(md1.signup_category)
```

# Over Sampling

```python
count_0, count_1 = md1.label.value_counts()
count_0
count_1
```

```python
md1_0 = md1[md1['label'] == 0]
md1_1 = md1[md1['label'] == 1]
md1_over = md1_1.sample(count_0, replace = True)
md1_over = pd.concat([md1_over,md1_0 ], axis = 0)
```

```python
md1_over_x = md_over.iloc[:, 0:13]     #independent variables
md1_over_y = md_over.iloc[:, -1]       #dependent varaiable
md1_over_x.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
md1_x_train,md1_x_test,md1_y_train,md1_y_test = train_test_split(md1_over_x,md1_over_y,test_size = .2, random_state = 101)
```
